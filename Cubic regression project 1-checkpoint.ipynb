```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in Cubic[cols].columns:
    lr,ur=remove_outlier(Cubic[column])
    Cubic[column]=np.where(Cubic[column]>ur,ur,Cubic[column])
    Cubic[column]=np.where(Cubic[column]<lr,lr,Cubic[column])
```

# Train Test split

```python
X = Cubic.drop('price', axis=1)
y = Cubic[['price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30 , random_state=1)
```

#Linear Regression model

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
# R square on training data
regression_model.score(X_train, y_train)
```

```python
# R square on testing data
regression_model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
#RMSE on Training data
predicted_train=regression_model.fit(X_train, y_train).predict(X_train)
np.sqrt(metrics.mean_squared_error(y_train,predicted_train))
```

```python
#RMSE on Testing data
predicted_test=regression_model.fit(X_train, y_train).predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test,predicted_test))
```

#Linear Regression using statsmodels(OLS)

```python
import statsmodels.api as sm
```

```python
X_train=sm.add_constant(X_train)
X_test=sm.add_constant(X_test)
```

```python
model = sm.OLS(y_train,X_train).fit()
model
```

```python
model.summary()
```

```python
#Calculate MSE on train
predicted_train=model.predict(X_train)
mse=metrics.mean_squared_error(y_train,predicted_train)
```

```python
np.sqrt(mse) #RMSE on train 
```

```python
# Prediction on Test data
y_pred = model.predict(X_test)
```

```python
#Calculate MSE on test
mse=metrics.mean_squared_error(y_test,y_pred)
np.sqrt(mse) #RMSE on test
```

```python
plt.scatter(y_test, y_pred)
plt.show()
```
